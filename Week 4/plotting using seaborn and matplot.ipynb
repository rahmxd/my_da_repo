```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

np.random.seed(42)
n_points = 100
x = np.random.rand(n_points)
y = 2*x+np.random.normal(0,0.2, n_points)

df = pd.DataFrame({'X': x, 'Y': y})
```

```python
df.plot(kind = 'scatter', x='X', y='Y')
```
